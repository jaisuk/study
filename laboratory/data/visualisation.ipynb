```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../../datasets/body.csv', index_col =0)
df.head()
```

```python
df['Height']
```

```python
df['Height'].value_counts().sort_index().plot()
```

```python
sns.kdeplot(df['Height'], bw=0.01)
```

```python
df.plot(kind='hist', y='Height')
```

```python
sns.distplot(df['Height'], bins=15)
```

```python
df.plot(kind='box', y='Height')
```

```python
sns.violinplot(y=df['Height'])
```

```python
df.plot(kind='scatter', x='Height', y='Weight')
```

```python
sns.kdeplot(df['Height'], df['Weight'])
```

```python
sns.lmplot(data=df, x='Height', y='Weight')
```
